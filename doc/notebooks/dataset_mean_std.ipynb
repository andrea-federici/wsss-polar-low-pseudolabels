```python
import sys

from pathlib import Path
root = Path('../..').resolve()
sys.path.insert(0, str(root))

from torchvision import datasets, transforms

from src.data.utils import compute_dataset_mean_std
```

```python
pl_train_dir = root / "data/polar_lows/train"
pl_image_size = (800, 800)

pl_train_data = datasets.ImageFolder(
    pl_train_dir, 
    transform=transforms.Compose([
        transforms.Resize(pl_image_size),
        transforms.ToTensor()
    ])
)

pl_mean, pl_std = compute_dataset_mean_std(pl_train_data)
print(f"POLARLOWS - Mean: {pl_mean}, Std: {pl_std}")
```

```python
c_train_dir = root / "data/bus_final/train"
c_image_size = (512, 512)

c_train_data = datasets.ImageFolder(
    c_train_dir, 
    transform=transforms.Compose([
        transforms.Resize(c_image_size),
        transforms.ToTensor()
    ])
)

c_mean, c_std = compute_dataset_mean_std(c_train_data)
print(f"CANCER - Mean: {c_mean}, Std: {c_std}")
```
